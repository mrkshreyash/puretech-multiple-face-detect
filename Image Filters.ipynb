```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
```

```python
img = Image.open(r"test_images\test 2.JPG")
plt.imshow(img)
```

## Contrast Adjustment in an Image

---

```python
enh = ImageEnhance.Contrast(img)
thirty_percent_contrast_img = enh.enhance(1.3)
sixty_percent_contrast_img = enh.enhance(2.6)
ninty_percent_contrast_img = enh.enhance(3.9)
```

```python
fig, axis = plt.subplots(2, 2)

axis[0][0].imshow(img)
axis[0][0].set_title("Without contrast")

axis[0][1].imshow(thirty_percent_contrast_img)
axis[0][1].set_title("With 30% contrast")

axis[1][0].imshow(sixty_percent_contrast_img)
axis[1][0].set_title("With 60% contrast")

axis[1][1].imshow(ninty_percent_contrast_img)
axis[1][1].set_title("With 90% contrast")

plt.tight_layout()
plt.show()
```
